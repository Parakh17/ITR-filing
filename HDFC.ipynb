```python
#removing stopping words from the description to use it for further labeling
displayN = df1["Description"]
displayN.head()
```

```python
lst = ["[0-9]",":","\n"]
for i in lst:
    displayN.replace(to_replace=f"{i}", value="", regex=True,inplace=True)
displayN.replace(to_replace=f"-", value=" ", regex=True,inplace=True)
displayN.replace(to_replace=f"/", value=" ", regex=True,inplace=True)
displayN.replace(to_replace=f"\*", value=" ", regex=True,inplace=True)
displayN.replace(to_replace=f"@", value=" ", regex=True,inplace=True)
```

```python
displayN
```

```python
#striping the description of empty spaces
cleanedSTR = displayN.astype({"Description":"string"})

for i in range(len(cleanedSTR)):
    x = cleanedSTR[i].split()
    for j in range(len(x)):
        x[j] = x[j].strip()
    cleanedSTR[i] = " ".join(x)
print(cleanedSTR)
```

## LABELLING THE DATA

```python
#creating labels
labels = {"imps": "imps",  "rrn": "imps", "loan": "loan", "emi": "emi", "amazon": "shopping", 
          "flipkart": "shopping", "mutualfund": "invest", "wallet":"paytm","upi": "upi",
          "txn paytm": "paytm", "restaurant": "food", "paytm": "paytm", "pytm": "paytm",
          "atd": "atm", "atm": "atm", "net txn": "nettxn", "cash": "cash", "funds trf": "trf",
          "neft": "neft", "interest": "interest", "debit": "debit", "atw":"atm",
          "metro": "travel", "swiggy": "food", "faasos": "food", "zomato": "food", "ola": "travel",
          "refund": "refund", "cash deposit": "cash deposit", "cash cheque": "cash cheque",
          "charge": "bank_charges", "pca": "trf"}
```

```python
#labelled data will be stored in this list
labelLST = []
```

```python
#labelling the rows
for i in cleanedSTR:
    flag = 0
    for j in list(labels.keys()):
        if j in i.lower():
            labelLST.append(labels[j])
            flag = 1
            break
    if flag == 0:
        labelLST.append("miscellaneous")
```

```python
#adding the labelled column to the dataframe
df1["label"] = pd.DataFrame(labelLST)
```

```python
#earlier most of the transactions were labelled as "upi"
#so i created a loop to give a proper labelling to the transactions and marking it with the proper names 
for i in range(len(df1)):
    if df1["label"].iloc[i] == "upi":
        try: 
            x = df1["Description"].iloc[i].split()[1]
            x += df1["Description"].iloc[i].split()[2]
        except IndexError:
            x = df1["Description"].iloc[i].split()[3]
        df1["label"].iloc[i] = x
```
